# DATA PREPROCESSING

#####  In a Gene column, there multiple genes in one row, so I need to separate them first

```python
new_df = df.head(10).copy()
new_df
```

```python
count = 0
for gen in df["Gene"]:
    if "," in str(gen):
        count += 1
print(count)
```

```python

for gen in df["Gene"]:
    list_of_genes = []
    if "," in str(gen):
        list_of_genes.append(gen)
    print(list_of_genes)
```

```python
result = {}

for key, value in zip(df["reporterID"], df["Gene"]):
    key = str(key)
    if key not in result:
        result[key] = []
    
    if "," in str(value):
        for gene in value.split(","):
            result[key].append(gene.strip())
    else:
        result[key].append(str(value).strip())
result
```

```python
df_res = pd.DataFrame([
    {"reporterID": key, "Gene": gene}
    for key, genes in result.items()
    for gene in genes
])
```

```python
df_res
```

```python
df_res["reporterID"] = df_res["reporterID"].astype(str)
df["reporterID"] = df["reporterID"].astype(str)
```

```python
merged_df = pd.merge(df_res, df, on="reporterID", how="left")

```

```python
merged_df.drop('Gene_y', axis=1, inplace=True)
```
